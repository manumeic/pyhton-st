##### Podemos decir entonces que los años 2017, 2018 y 2019 tienen las siguientes columnas con la siguiente información:

```python
#Unimos los 3 datasets, podemos hacerlo asi directamente ya que tienen las mismas columnas
dataset = pd.concat([data_2019, data_2018, data_2017])

#Nos quedamos solamente con la estación Illia y la forma de pago Efectivo
mask = np.logical_and(dataset.estacion=='Illia', dataset.forma_pago == 'EFECTIVO')
dataset = dataset[mask]

#Quitamos las columnas que no vamos a utilizar
dataset.drop(columns = ['periodo','hora_inicio','forma_pago', 'estacion'], inplace = True)

dataset.head()
```

```python
diario_sin_outliers['dia'] = diario_sin_outliers['fecha'].dt.day
```

```python
fecha_corte = '2019-10-01'

train = diario_sin_outliers[diario_sin_outliers.fecha < fecha_corte]
test = diario_sin_outliers[diario_sin_outliers.fecha >= fecha_corte]

X_train = train[['dia']]
y_train = train [['cantidad_pasos']]

X_test = test[['dia']]
y_test = test[['cantidad_pasos']]
```

```python
rl = LinearRegression()

rl.fit(X_train, y_train)
```

```python
rl_train = rl.predict(X_train)
rl_test = rl.predict(X_test)
```

```python
rmse_rltrain = np.sqrt(mean_squared_error(y_train, rl_train))
r2_rltrain = r2_score(y_train, rl_train)

rmse_rltest = np.sqrt(mean_squared_error(y_test, rl_test))
r2_rltest = r2_score(y_test, rl_test)
print('En el caso de las métricas de evaluación, para el set de entrenamiento el Error cuadrático medio es de', rmse_rltrain,
     'y el R2 es de', r2_rltrain,'.\nPara el set de entranamiento, el RMSE es de ',rmse_rltest, 'y el R2 es de', r2_rltest,'\nDe esto se desprende como conclusión que este modelo no es adecuado para predecir en este set de datos.')
```

```python
plt.figure(figsize = (20,10))
plt.plot(diario_sin_outliers.fecha, diario_sin_outliers.cantidad_pasos, label = 'Valores reales')
plt.plot(train.fecha, rl_train, label = 'Valores predichos en train')
plt.plot(test.fecha, rl_test, label = 'Valores predichos en test')
plt.legend()
plt.show()
```

```python
back = 30 #Cantidad de días que el modelo va a usar para predecir el siguiente valor
cantidad_pasos = diario_sin_outliers.cantidad_pasos
```

```python
N = cantidad_pasos.shape[0]
X = np.zeros((N - back - 1, back))
y = np.zeros(N - back - 1)
```

```python
#Relleno las matrices y separo en train y test teniendo en cuenta la fecha de corte 

for i in range(X.shape[0]):
    X[i,:] = cantidad_pasos.iloc[i:i+back]
    y[i] = cantidad_pasos.iloc[i+back]
    
X_train2 = X[:1003,:]
y_train2 = y[:1003]

X_test2 = X[1003:,:]
y_test2 = y[1003:]
```

```python
X_test2.shape
```

```python
rf = RandomForestRegressor(n_estimators = 75, random_state = 42)

rf.fit(X_train2, y_train2)
```

```python
rf_train = rf.predict(X_train2)
rf_test = rf.predict(X_test2)

rmse_rftrain = np.sqrt(mean_squared_error(y_train2, rf_train))
rmse_rftest = np.sqrt(mean_squared_error(y_test2, rf_test))
r2_rftrain = r2_score(y_train2, rf_train)
r2_rftest = r2_score(y_test2, rf_test)

print('Métricas del set de entrenamiento:\nRMSE:',rmse_rftrain,'\nR2:' ,r2_rftrain,'\n\nMétricas del set de test:\nRMSE:', rmse_rftest, '\nR2:', r2_rftest)
```

```python
randomf = RandomForestRegressor()

parametros = {'n_estimators' : [10, 50, 90, 150],
             'max_depth' : np.arange(8,15),
             'bootstrap' : [True,False]}

random_search = RandomizedSearchCV(estimator = randomf, param_distributions = parametros, n_iter = 10, cv = 5, random_state = 42)
```

```python
random_search.fit(X_train2, y_train2)
```

```python
print('Los mejores hiperparámetros que surgen de la búsqueda son:', random_search.best_params_)
```

```python
random_opt = RandomForestRegressor(n_estimators = 50, max_depth = 9, bootstrap = True)
random_opt.fit(X_train2, y_train2)
```

```python
random_train = random_opt.predict(X_train2)
random_test = random_opt.predict(X_test2)

rmse_randomtrain = np.sqrt(mean_squared_error(y_train2, random_train))
rmse_randomtest = np.sqrt(mean_squared_error(y_test2, random_test))
r2_randomtrain = r2_score(y_train2, random_train)
r2_randomtest = r2_score(y_test2, random_test)

print('Métricas del set de entrenamiento:\nRMSE:',rmse_randomtrain,'\nR2:' ,r2_randomtrain,'\n\nMétricas del set de test:\nRMSE:', rmse_randomtest, '\nR2:', r2_randomtest)
```
